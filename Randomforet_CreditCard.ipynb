```python
df.rename(columns={"default.payment.next.month": "default_payment"},inplace=True)
```

```python
df.info()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest, ytrain,ytest = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
Xtrain.shape, ytrain.shape
```

```python
Xtest.shape, ytest.shape
```

```python
num_cols = X.select_dtypes(include='number').columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Numerical Pipeline ------------------------------------------------------------------------------------------------------------

num_pipe = Pipeline(steps=[
    ('impute',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])

# Column Tansformer -------------------------------------------------------------------------------------------------------------

preprocessing = ColumnTransformer([
    ('numerical_pipe',num_pipe, num_cols)
])
```

```python
scaled_xtrain = preprocessing.fit_transform(Xtrain)
```

```python
scaled_xtest = preprocessing.transform(Xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier()
```

```python
model.fit(scaled_xtrain,ytrain)
```

```python
ypred = model.predict(scaled_xtest)
```

```python
accuracy_score(ytest,ypred)
```

```python
from sklearn.metrics import confusion_matrix

print(f"Confusion Metrics \n {confusion_matrix(ytest,ypred)}")
print("---------------------------------------------------------------------------------")

sns.heatmap(data = confusion_matrix(ytest,ypred),annot=True)
plt.show()
```
