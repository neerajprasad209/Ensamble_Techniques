# Numerical Pipeline ______________________________________________________________________________________________________________________

    ('impute',SimpleImputer(strategy='mean')),

Model Tranning  </H1>

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
model.fit(Xtrain, ytrain)
```

```python
ypred = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest,ypred)
```

```python
## Hypeparameter Tuning
params={'max_depth':[3,5,10,None],
            'n_estimators':[100,200,300],
            'criterion':['gini','entropy']
            }
```

```python
classifer = RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV

cv = RandomizedSearchCV(classifer, param_distributions=params, scoring='accuracy', cv=5, verbose=3)
cv.fit(Xtrain,ytrain)
```

```python
cv.best_params_
```

```python
cv.best_estimator_
```

```python
y_pred = cv.predict(Xtest)
```

```python
accuracy_score(ytest,y_pred)
```
