Seperating the Independent and Dependent Variable </H1>

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

Perfoming the Train Test Split </H1>

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest, ytrain,ytest = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
Xtrain.shape, ytrain.shape
```

```python
Xtest.shape, ytest.shape
```

<H1 style = "font-size: 25px;

Decision Tree Model Tranning  </H1>

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=6)
```

```python
tree.fit(Xtrain,ytrain)
```

```python
ypred = tree.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("------------------------------------Accuracy Score:-------------------------------------------------")
print(f"\n{accuracy_score(ytest,ypred)*100}")
print("------------------------------------Calssification Report:-------------------------------------------------")
print(f"\n{classification_report(ytest,ypred)}")
print("------------------------------------Confusion Metrix:-------------------------------------------------")
print(f"\n{confusion_matrix(ytest,ypred)}")
```

Random Forest Model Tranning  </H1>

```python
## Hypeparameter Tuning
params={'max_depth':[2,3,4,5,6,7,10,None],
        'n_estimators':[10,20,30],
        'criterion':['gini','entropy','log_loss']
        }
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

model = RandomForestClassifier()

clf = GridSearchCV(model, param_grid= params, scoring='accuracy', cv=3, verbose=3)
```

```python
clf.fit(Xtrain,ytrain)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("------------------------------------Accuracy Score:-------------------------------------------------")
print(f"\n{accuracy_score(ytest,y_pred)*100}")
print("------------------------------------Calssification Report:-------------------------------------------------")
print(f"\n{classification_report(ytest,y_pred)}")
print("------------------------------------Confusion Metrix:-------------------------------------------------")
print(f"\n{confusion_matrix(ytest,y_pred)}")
```
